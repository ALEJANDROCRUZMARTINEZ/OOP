## Clase Employee
- **int id**: Atributo de instancia. Identifica de manera unica a cada instancia de `Employee`.
- **int new_id**: Atributo de clase. Identificador unico.
- **void say_id()**: Imprime 'Mi id es: {self.id}.
    


```python
class Employee:
    new_id= 1

    def __init__(self):
        self.id =Employee.new_id
        Employee.new_id += 1

    def say_id(self):
        print(f' Mi id es: {self.id}')
```

```python
Tania = Employee()
Tania.say_id()
```

## Clase User
- **str role**: Atributo de instancia. Funcion de `User` p.ej.
    - Administrador
    - Manager
    - Usuario
- **str username**: Atributo de instancia. Nombre del usuario.
- **void say_user_info()**: Imprime 'Mi nombre es {self.username} y soy {self.role}.'.

```python
class User:
    def __init__(self, role, username):
        self.role = role
        self.username = username
    
    def say_user_info(self):
        print(f'Mi nombre es {self.username} y soy {self.role}.')
```

```python
Juan = User("Administrador", "Juan")
Juan.say_user_info()  
Maria = User("Usuario", "María")
Maria.say_user_info() 
Marta = User("Manager", "Marta")
Marta.say_user_info() 
```

## Clase Admin
- **void say_id()**: Imprime:
    - 'Mi id es: {self.id}.
    - 'Soy un Administrador \\(0-0)/'.

```python
class Admin(Employee, User):
    def __init__(self, username):
        super().__init__()
        User.__init__(self, 'administrador', username)

    def say_id(self):
        super().say_id()
        print(f'Soy un {self.role}\(0-0)/')
```

```python
Jose = Admin('jose')
Jose.say_id()
Jose.say_user_info()
```

## Clase Manager
- **void say_id()**: Imprime:
    - 'Mi id es: {self.id}.
    - 'Soy un Administrador \\(0-0)/'.    
    - 'Y estoy al mando'

```python
class Manager(Admin):
    def say_id(self):
        super().say_id()
        print(' Y estoy al mando')
```

```python
Cesar = Manager()
Cesar.say_id()
```
